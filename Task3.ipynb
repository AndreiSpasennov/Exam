```python
import numpy as np
import pandas as pd
```

```python
attend = np.array([0.63, 0.76, 0.94, 0.23, 0.48])
mark = np.array([70, 74, 89, 35, 68])
name = ['Mark', 'Liz', 'Ben', 'Tony', 'Mary']
data = {"Attendance": attend, "Marks": mark}
df = pd.DataFrame(data, index=name)
```

```python
import matplotlib.pyplot as plt
plt.bar(df["Marks"], df["Attendance"])
plt.show
```
